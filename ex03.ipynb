```python
df.dropna(how="all", axis=1, inplace=True)
```

### Removendo candidatos que não tem e-mail ou composição legenda

```python
df.dropna(axis=0, how='any', subset=['composicao_legenda', 'email'])
```

### Filtrando

```python
df['descricao_cargo'].unique()
```

```python
df_presidente_PT = df[ ( df ['descricao_cargo'] == 'PRESIDENTE') & 
                ( df ['sigla_partido'] == 'PT')] 
                
```

```python
df_presidente = df[ (df['descricao_cargo'] == 'PRESIDENTE')].copy()
```

### Pegando candidatos a presidencia

```python
df_presidente.columns
```

```python
colunas_interesse =  ['ano_eleicao',
                     'numero_turno',
                     'cpf',
                     'data_nascimento',
                     'descricao_cor_raca',
                     'descricao_estado_civil',
                     'descricao_genero',
                     'descricao_grau_instrucao',
                     'descricao_ocupacao',
                     'email',
                     'nome',
                     'nome_social',
                     'sigla_uf_nascimento',
                     'sigla_partido',
                     'descricao_cargo',
                     'descricao_detalhe_situacao_candidatura']
```

```python
df_presidente = df_presidente[colunas_interesse]
```

```python
df_presidente['cpf'].nunique()
```

```python
df_presidente.shape
```

### Removendo cpf repetido

```python
df_presidente.sort_values(by=['cpf', 'numero_turno']).drop_duplicates(subset=['cpf'], keep='first')
```

### Removendo candidatos INAPTOS

```python
df_presidente.where(df_presidente['descricao_detalhe_situacao_candidatura'] == "APTO").dropna(how="any")
```
