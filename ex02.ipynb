```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('./data/tb_candidatura_2018.csv', sep=';')
df.head()
```

```python
tipos_colunas = df[['idade_data_eleicao', 'idade_data_posse']].dtypes
type(tipos_colunas)
print(tipos_colunas)
```

```python
df[['idade_data_eleicao', 'idade_data_posse']].describe()
```

```python
df_novo = df.dropna(how='all', axis=1).copy()
```

```python
df_novo.head()
```

```python
df_novo['idade_data_posse'] = df_novo['idade_data_posse'].astype(int)
```

```python
df_novo['idade_data_posse'].dtypes
```

#### Calculando log da idade

```python
df_novo['idade_data_posse_log'] = np.log(df_novo['idade_data_posse'])
```

```python
df_novo[['idade_data_posse','idade_data_posse_log' ]].head()
```

#### Testando novas formulas

```python
df_novo['campo_maluco'] = (df_novo['idade_data_posse'] + df_novo['idade_data_posse_log']) * 2
```

```python
df_novo[['idade_data_posse','idade_data_posse_log', 'campo_maluco' ]].head()
```

```python
def pega_primeiro_nome(nome:str):
    return nome.lstrip(' ').split(' ')[0]
```
